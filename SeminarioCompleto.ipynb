```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
##Creamos columnas para nuestro dataframe
cols=('Grupo', 'Total1995', 'Hombres1995', 'Mujeres1995' , 'Total2000', 'Hombres2000', 'Mujeres2000', 'Total2005', 'Hombres2005', 'Mujeres2005', 'Total2010', 'Hombres2010', 'Mujeres2010', 'Total2020', 'Hombres2020', 'Mujeres2020')
```

```python
##Leemos el archivo csv
df=pd.read_csv('/home/kevin/Descargas/PoblacionL.csv', names=cols)
df
```

```python
##Creamos una lista con los grupos de edad de mas de 65 años
df['Total1995'].iloc[16:24].astype(float)
```

```python
##Sumamos los valores de la lista
Ed1995= df['Total1995'].iloc[16:24].astype(float).sum()
Ed1995
```

```python
##creamos un espacio para guardar los resultados
resultados = {}
```

```python
##Creamos un bucle para sumar los valores de cada columna de las personas mayores de 65 años
for col in df.columns:
    if col != 'Grupo':
        suma = df.loc[16:23, col].astype(float).sum()
        resultados[col] = suma
```

```python
##Creamos un dataframe con los resultados
resultados_df = pd.DataFrame(resultados, index=[0])
```

```python
resultados_df
```

```python
##Transponemos el dataframe
resultados_df = resultados_df.T
```

```python
##Reseteamos el indice
resultados_df= resultados_df.reset_index()
```

```python
resultados_df
```

```python
##Nombramos las columnas
resultados_df.columns = ['Año', 'Poblacion']
```

```python
resultados_df['año'] = resultados_df['Año'].str[-4:]

resultados_df['genero'] = resultados_df['Año'].str[:-4]
```

```python
##Eliminamos la columna Año que no nos sirve
resultados_df= resultados_df.drop('Año', axis=1)
```

```python
resultados_df
```

```python
##Graficamos los resultados
df_agrupado = resultados_df.groupby(['año', 'genero']).sum()

# Creamos una lista de los géneros para usar como índice en la siguiente fase
generos = ['Total', 'Hombres', 'Mujeres']

# Creamos un dataframe vacío para almacenar los datos agrupados por género
df_generos = pd.DataFrame(index=resultados_df['año'].unique(), columns=generos)

# Rellenamos el dataframe con los datos agrupados por género
for g in generos:
    df_generos[g] = df_agrupado.loc[(slice(None), g), 'Poblacion'].values

# Creamos la gráfica
df_generos.plot(kind='bar')
plt.title('Población de la tercera edad por género y año')
plt.xlabel('Año')
plt.ylabel('Población')
plt.show()
```

```python
totales = df.iloc[[2],[1,4,7,10,13]]
totales
```

```python
##Cambiamos el tipo de dato de las columnas
totales=totales.astype({"Total1995" : "float64", "Total2000" : "float64", "Total2005" : "float64", "Total2010" : "float64", "Total2020" : "float64" })
```

```python
totales
```

```python
## Usaremos como hipótesis que la tasa para el incremento poblacional es la misma durante el quinquenio en curso
inc_2010 = (totales["Total2020"]-totales["Total2010"])/5
inc_2005 = (totales["Total2010"]-totales["Total2005"])/5
inc_2000 = (totales["Total2005"]-totales["Total2000"])/5
inc_1995 = (totales["Total2000"]-totales["Total1995"])/5
## Creamos una lista con los años desde 1995 hasta 2020
incrementos = [inc_1995,inc_2000,inc_2005,inc_2010]
```

```python
incrementos
```

```python
##Creamos un dataframe con los incrementos
incrementos_df = pd.DataFrame(incrementos, index=[1995,2000,2005,2010])
```

```python
##Reseteamos el indice
incrementos_df=incrementos_df.reset_index()
```

```python
##Nombramos las columnas
incrementos_df.columns = ['Año', 'Incremento']
```

```python
incrementos_df
```

```python
## Creamos otro dataframe para ingresar los datos
columnas = ["Año", "Total"]
poblacion_por_años = pd.DataFrame(columns=columnas)
## Creamos una lista con los años desde 1995 hasta 2020
año = []
y = 1995
while y <= 2020:
    año.append(y)
    y += 1
## Obtenemos los datos con el supuesto de que la tasa para el incremento poblacional es la misma durante el quinquenio en curso
pob = int(totales["Total1995"])
datos_pob = [pob]
for incremento in incrementos_df['Incremento']:
    n = 0
    while n<5:
        pob += incremento
        round(pob,2)
        datos_pob.append(pob)
        n+=1
x=0
while x<5:
    pob += inc_2010
    round(pob, 2)
    datos_pob.append(pob)
    x += 1
## Redondeamos los datos
datos_pob = [round(año) for año in datos_pob]
```

```python
## Llenamos el dataframe con los datos
poblacion_por_años["Año"] = año
poblacion_por_años["Total"] = datos_pob
```

```python
poblacion_por_años['Total'] = poblacion_por_años['Total'].replace('dtype: float', '').astype(float)
poblacion_por_años
```

```python
## Unimos los dataframe
poblacion_mortalidad = pd.merge(Mor,poblacion_por_años, on = "Año", how = "inner")
## Obtenemos la tasa de mortalidad por año
poblacion_mortalidad["Tasa_de_mortalidad"] = poblacion_mortalidad["Defunciones registradas"]/poblacion_mortalidad["Total"]
poblacion_mortalidad
```

```python
##Graficamos los resultados
poblacion_mortalidad.plot(kind='bar', x='Año', y='Tasa_de_mortalidad', color='purple', figsize=(10, 5))
plt.title('Tasa de mortalidad por año')
plt.xlabel('Año')
plt.ylabel('Tasa de mortalidad')
plt.show()
```

```python
##Graficamos los resultados para ver tendencia
fig, ax = plt.subplots()

# Graficar la tasa de mortalidad por año
ax.plot(poblacion_mortalidad['Año'], poblacion_mortalidad['Tasa_de_mortalidad'])

# Agregar etiquetas y título
ax.set_xlabel('Año')
ax.set_ylabel('Tasa de mortalidad')
ax.set_title('Tasa de mortalidad por año')

# Mostrar la gráfica
plt.show()
```
